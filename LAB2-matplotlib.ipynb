```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_ekseni = np.arange(0,20,0.5)
y_ekseni = np.tan(x_ekseni)
plt.plot(x_ekseni , y_ekseni)
plt.title("Tanjant Grafiği")
plt.xlabel("X Ekseni")
plt.ylabel("Y Ekseni")


plt.show()
```

```python
#Scatter Plot: iki farklı değer arasındaki ilişkiyi belirlemek için kullanılan ve noktalardan oluşan bir tablodur.
x= np.random.rand(20)
y = np.random.rand(20)
renk = np.random.rand(20)
boyut = np.random.randint(50,150,20)

plt.scatter(x,y,c=renk,s=boyut)
plt.title("Nokta Gösterimi")
plt.xlabel("X Ekseni")
plt.ylabel("Y Ekseni")


plt.show()
```

```python
# Verileri Histogram ile İfade Edersek

veri=np.random.randn(1000) #random sayı üretiyoruz
plt.hist(veri, color="yellow" , bins=50) #bins histogramdaki kutu sayısı
plt.title("Histogram")
plt.xlabel("Değerler")
plt.ylabel("Sıklık")

# Show graphics
plt.show()
```

```python
# Data
film = ['Korku', 'Gerilim', 'Komedi', 'Kara Mizah', 'Bilim-Kurgu']
deger = np.random.randint(1, 100, 5)

# Graph creation
plt.bar(film,deger)

# Determine chart properties
plt.title("Sütun Grafiği")
plt.xlabel("Film Türleri")
plt.ylabel("İzlenme oranı(%)")

# Show graphics
plt.show()
```

```python
oran = [15,20,25,30,35,40,50]

plt.pie(oran)

plt.title("Pasta Dilimi")

plt.show()
```

```python
#3D şeklinde göstermek için axes3D kütüphanesini eklemeliyiz
from mpl_toolkits.mplot3d import Axes3D
```

```python
x_ekseni = np.arange(0,5,0.25)
y_ekseni = np.arange(0,5,0.25)
x_ekseni,y_ekseni = np.meshgrid(x_ekseni , y_ekseni) #meshgrid ile matris görünümü elde edilir
r = np.sqrt(x_ekseni**2 + y_ekseni**2)
z = np.sin(r)


figure = plt.figure()
ax = figure.add_subplot(111, projection='3d')


ax.plot_surface(x_ekseni, y_ekseni, z)

# Determine chart properties
ax.set_title("3D Grafik")
ax.set_xlabel("X Ekseni")
ax.set_ylabel("Y Ekseni")
ax.set_zlabel("Z Ekseni")

# Show
plt.show()
```
